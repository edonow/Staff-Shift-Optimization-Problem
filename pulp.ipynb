## Problem 1

Solve for x and y satisfying the following conditions

- 120x + 150y = 1440
- x + y = 10


```python
problem = LpProblem("TestProblem", LpMaximize)

x = LpVariable("x", cat=LpContinuous, lowBound=0)
y = LpVariable("y", cat=LpContinuous, lowBound=0)

problem += 120 * x + 150 * y == 1440
problem += x + y == 10

status = problem.solve()

print(f"status: {LpStatus[status]}")
print(f"x: {x.varValue}")
print(f"y: {y.varValue}")
```

## Problem 2

Solve for x and y satisfying the following conditions

- maxiumize:

  - 2x + 3y

- condition:
  - x + 3y <=9
  - x + y <= 4
  - 2x + y <=6


```python
problem = LpProblem("TestProblem", LpMaximize)

x = LpVariable("x", cat=LpContinuous, lowBound=0)
y = LpVariable("y", cat=LpContinuous, lowBound=0)

problem += 1 * x + 3 * y <= 9
problem += 1 * x + 1 * y <= 4
problem += 2 * x + 1 * y <= 6
problem += 2 * x + 3 * y


status = problem.solve()

print(f"status: {LpStatus[status]}")
print(f"x: {x.varValue}")
print(f"y: {y.varValue}")
print(f"objective: {pulp.value(problem.objective)}")
```

## Problem 4

From the table below, please maximize profits within a limited number of inventory items.
| | Juice A | Juice B | Juice C | Stock |
| --------- | ------- | ------- | ------- | ----- |
| Fruit A | 1 | 2 | 3 | 60 |
| Fruit B | 3 | 0 | 1 | 36 |
| Fruit C | 2 | 4 | 0 | 48 |
| Gain(JPY) | 150 | 200 | 300 | |


```python
problem = LpProblem("TestProblem", LpMaximize)
x1 = LpVariable("x1", cat=LpContinuous, lowBound=0)
x2 = LpVariable("x2", cat=LpContinuous, lowBound=0)
x3 = LpVariable("x3", cat=LpContinuous, lowBound=0)

problem += 1 * x1 + 2 * x2 + 3 * x3 <= 60
problem += 3 * x1 + 0 * x2 + 1 * x3 <= 36
problem += 2 * x1 + 4 * x2 + 0 * x3 <= 48
problem += 150 * x1 + 200 * x2 + 300 * x3

status = problem.solve()
print(f"status: {LpStatus[status]}")
print(f"x1: {x1.varValue}")
print(f"x2: {x2.varValue}")
print(f"x3: {x3.varValue}")
print(f"objective: {value(problem.objective)}")
```

## Problem 5

From the table below, please maximize profits within a limited number of inventory items.
| | Veggie A | Veggie B | Veggie C | Required |
| ---------: | -------: | -------: | -------: | -----: |
| Feature A | 10 | 25 | 30 | 50 |
| Feature B | 15 | 5 | 35 | 60 |
| Feature C | 2 | 2 | 20 | 10 |
| Feature D | 5 | 80 | 40 | 40 |
| Cost(JPY) | 400 | 250 | 1000 | |


```python
problem = LpProblem("TestProblem", LpMinimize)
x1 = LpVariable("x1", cat=LpContinuous, lowBound=0)
x2 = LpVariable("x2", cat=LpContinuous, lowBound=0)
x3 = LpVariable("x3", cat=LpContinuous, lowBound=0)

problem += 10 * x1 + 25 * x2 + 30 * x3 >= 50
problem += 15 * x1 + 5 * x2 + 35 * x3 >= 60
problem += 2 * x1 + 2 * x2 + 20 * x3 >= 10
problem += 5 * x1 + 80 * x2 + 40 * x3 >= 40
problem += 400 * x1 + 250 * x2 + 1000 * x3

status = problem.solve()
print(f"status: {LpStatus[status]}")
print(f"x1: {x1.varValue:.4f}")
print(f"x2: {x2.varValue:.4f}")
print(f"x3: {x3.varValue:.4f}")
print(f"objective: {pulp.value(problem.objective):.4f}")
```

### other solution


```python
required = {"feature_a": 50, "feature_b": 60, "feature_c": 10, "feature_d": 40}
cost = {"veggie_a": 400, "veggie_b": 250, "veggie_c": 1000}
matrix = {
    ("feature_a", "veggie_a"): 10,
    ("feature_a", "veggie_b"): 25,
    ("feature_a", "veggie_c"): 30,
    ("feature_b", "veggie_a"): 15,
    ("feature_b", "veggie_b"): 5,
    ("feature_b", "veggie_c"): 35,
    ("feature_c", "veggie_a"): 2,
    ("feature_c", "veggie_b"): 2,
    ("feature_c", "veggie_c"): 20,
    ("feature_d", "veggie_a"): 5,
    ("feature_d", "veggie_b"): 80,
    ("feature_d", "veggie_c"): 40,
}
F = list(required.keys())
V = list(cost.keys())

problem = LpProblem("TestProblem4", LpMinimize)
x = LpVariable.dicts("x", V, cat=LpContinuous, lowBound=0)

for f in F:
    problem += lpSum(matrix[f, v] * x[v] for v in V) >= required[f]
problem += lpSum(cost[v] * x[v] for v in V)

status = problem.solve()

print(f"Status: {LpStatus[status]}")
for v in V:
    print(f"{v}: {x[v].varValue:.4f}")
print(f"Objective value: {value(problem.objective):.4f}")
```

## Problem 6: Optimization of production volume

From the table below, please maximize profits within a limited number of inventory items.

|           |  p1 |  p2 |  p3 |  p4 | Stock |
| --------: | --: | --: | --: | --: | ----: |
|        m1 |   2 |   3 |   0 |   2 |    35 |
|        m2 |   0 |   2 |   2 |   2 |    22 |
|        m3 |   1 |   0 |   2 |   2 |    27 |
| Gain(JPY) |   3 |   4 |   4 |   5 |       |


```python
gain_df = pd.read_csv("./data/gains.csv")
requires_df = pd.read_csv("./data/requires.csv")
stocks_df = pd.read_csv("./data/stocks.csv")

P = requires_df["p"].unique()
M = requires_df["m"].unique()

gain = {row.p: row.gain for row in gain_df.itertuples()}
requires = {(row.p, row.m): row.require for row in requires_df.itertuples()}
stocks = {row.m: row.stock for row in stocks_df.itertuples()}


problem = LpProblem("TestProblem", LpMaximize)
x = LpVariable.dict("x", P, cat=LpInteger, lowBound=0)

for m in M:
    problem += lpSum(requires[p, m] * x[p] for p in P) <= stocks[m]

problem += lpSum(gain[p] * x[p] for p in P)

status = problem.solve()
print(f"status: {LpStatus[status]}")
```

```python
for p in P:
    print(f"x[{p}]: {x[p].varValue}")
print(f"objective: {pulp.value(problem.objective)}")
```

## Problem 7: Classroom Assignment Problem

Define constraints:

1. **A student is assigned to exactly one class**:

   - Each student `s` must be assigned to exactly one class `c`.

2. **Each class must have between 39 and 40 students**:

   - Each class `c` must have at least 39 students and at most 40 students.

3. **The number of male students in each class is at most 20**:

   - The number of male students in each class `c` must be no more than 20.

4. **The number of female students in each class is at most 20**:

   - The number of female students in each class `c` must be no more than 20.

5. **The average score of each class must be within ±10 of the overall mean score**:

   - The average score of students in each class `c` must be within ±10 of the overall mean score.

6. **Each class must have at least 2 leaders**:

   - Each class `c` must have at least 2 student leaders assigned.

7. **Each class can have at most 1 support student**:

   - Each class `c` can have at most 1 support student assigned.

8. **Certain pairs of students cannot be assigned to the same class**:

   - Certain pairs of students `s1` and `s2` must not be assigned to the same class.

9. **Initial class assignment should be respected (with penalties)**:
   - The initial class assignment should be respected according to `init_flag`, influencing the final allocation.

For data, please refer to [https://www.ohmsha.co.jp/book/9784274231759/](https://www.ohmsha.co.jp/book/9784274231759/).


```python
students_df = pd.read_csv("./data/students.csv")
pairs_df = pd.read_csv("./data/student_pairs.csv")

display(students_df.head(3), pairs_df.head(3))
display(students_df.shape, pairs_df.shape)
```

```python
S = students_df["student_id"].unique()
C = ["A", "B", "C", "D", "E", "F", "G", "H"]
SC = [(s, c) for s in S for c in C]
S_male = [row.student_id for row in students_df.itertuples() if row.gender == 1]
S_female = [row.student_id for row in students_df.itertuples() if row.gender == 0]
S_leader = [row.student_id for row in students_df.itertuples() if row.leader_flag == 1]
S_support = [row.student_id for row in students_df.itertuples() if row.support_flag == 1]
SS = [(row.student_id1, row.student_id2) for row in pairs_df.itertuples()]

S_scores = {row.student_id: row.score for row in students_df.itertuples()}
score_mean = students_df["score"].mean()

problem = LpProblem("ClassrommAsignmentProblem", LpMaximize)
x = LpVariable.dicts("x", SC, cat=LpBinary)

for s in S:
    problem += lpSum(x[s, c] for c in C) == 1

for c in C:
    problem += lpSum(x[s, c] for s in S) >= 39
    problem += lpSum(x[s, c] for s in S) <= 40
    problem += lpSum(x[s, c] for s in S_male) <= 20
    problem += lpSum(x[s, c] for s in S_female) <= 20
    problem += lpSum(x[s, c] for s in S_leader) >= 2
    problem += lpSum(x[s, c] for s in S_support) <= 1
for c in C:
    problem += (score_mean - 10) * lpSum(x[s, c] for s in S) <= lpSum(S_scores[s] * x[s, c] for s in S)
    problem += lpSum(S_scores[s] * x[s, c] for s in S) <= (score_mean + 10) * lpSum(x[s, c] for s in S)

for s1, s2 in SS:
    for c in C:
        problem += x[s1, c] + x[s2, c] <= 1


# Minimize variance of scores per class
# problem += -lpSum((S_scores[s] - score_mean) ** 2 * x[s, c] for s in S for c in C)


status = problem.solve()
print(f"status: {LpStatus[status]}")
```

```python
result_df = students_df.copy()
C2S = {}
S2C = {}
for c in C:
    C2S[c] = [s for s in S if x[s, c].varValue == 1]

for c, s in C2S.items():
    print(c, s)


S2C = {s: c for s in S for c in C if x[s, c].varValue == 1}
result_df["assigned_class"] = result_df["student_id"].map(S2C)


fig, axis = plt.subplots(4, 2, figsize=(12, 8))
axis = axis.flatten()
for i, c in enumerate(C):
    sns.histplot(result_df.query("assigned_class == @c")["score"], bins=range(0, 500, 40), kde=True, ax=axis[i])
    axis[i].set_title(f"Class {c}")
    axis[i].set_xlim(0, 500)
    axis[i].set_ylim(0, 20)
plt.tight_layout()
plt.show()
```

Re-optimize some classes with skewed grades


```python
new_C = {0: "A", 1: "B", 2: "C", 3: "D", 4: "E", 5: "F", 6: "G", 7: "H"}
students_df["socre_rank"] = students_df["score"].rank(ascending=False, method="first")
students_df["init_assigned_class"] = (students_df["socre_rank"] % len(C)).map(new_C)
init_class_flg = {(s, c): 0 for s in S for c in C}
for row in students_df.itertuples():
    init_class_flg[row.student_id, row.init_assigned_class] = 1


problem = LpProblem("ClassrommAsignmentProblem", LpMaximize)
x = LpVariable.dicts("x", SC, cat=LpBinary)

for s in S:
    problem += lpSum(x[s, c] for c in C) == 1

for c in C:
    problem += lpSum(x[s, c] for s in S) >= 39
    problem += lpSum(x[s, c] for s in S) <= 40
    problem += lpSum(x[s, c] for s in S_male) <= 20
    problem += lpSum(x[s, c] for s in S_female) <= 20
    problem += lpSum(x[s, c] for s in S_leader) >= 2
    problem += lpSum(x[s, c] for s in S_support) <= 1
for c in C:
    problem += (score_mean - 10) * lpSum(x[s, c] for s in S) <= lpSum(S_scores[s] * x[s, c] for s in S)
    problem += lpSum(S_scores[s] * x[s, c] for s in S) <= (score_mean + 10) * lpSum(x[s, c] for s in S)

for s1, s2 in SS:
    for c in C:
        problem += x[s1, c] + x[s2, c] <= 1

problem += lpSum(init_class_flg[s, c] * x[s, c] for s, c in SC)

status = problem.solve()
print(f"status: {LpStatus[status]}")
```

```python
result_df = students_df.copy()
C2S = {}
S2C = {}
for c in C:
    C2S[c] = [s for s in S if x[s, c].varValue == 1]

for c, s in C2S.items():
    print(c, s)


S2C = {s: c for s in S for c in C if x[s, c].varValue == 1}
result_df["assigned_class"] = result_df["student_id"].map(S2C)

fig, axis = plt.subplots(4, 2, figsize=(12, 8))
axis = axis.flatten()
for i, c in enumerate(C):
    sns.histplot(result_df.query("assigned_class == @c")["score"], bins=range(0, 500, 40), kde=True, ax=axis[i])
    axis[i].set_title(f"class: {c}")
    axis[i].set_xlim(0, 500)
    axis[i].set_ylim(0, 15)
plt.tight_layout()
plt.show()
```

## Problem 8: Classroom Assignment Problem

Define constraints:

1. **A student is assigned to exactly one class**:

2. **Each student has a high degree of hope**:

For data, please refer to [https://www.asakura.co.jp/detail.php?book_code=12895&srsltid=AfmBOooZf3qAcTWkcsc54k_UvKJmxssB493pVVIuzOEYcZ28pM0CD7ED](https://www.asakura.co.jp/detail.php?book_code=12895&srsltid=AfmBOooZf3qAcTWkcsc54k_UvKJmxssB493pVVIuzOEYcZ28pM0CD7ED).


```python
data_df = pd.read_excel("./data/data.xlsx")


data_df = data_df.fillna(0)
data_df = data_df.set_index("Unnamed: 0")
display(data_df.head(), data_df.shape)
```

- Check that each student is correctly ranked in the class of his/her choice
- Students rank 1, 2, 3, and 4 in any class.


```python
for i in range(data_df.shape[0]):
    t = data_df.iloc[i].values.sum()
    if abs(t - 10) > 1e-5:
        print(i, t)
        print(data_df.iloc[i].values)
```

```python
S = data_df.index.unique()
C = data_df.columns.unique()
SC = [(s, c) for s in S for c in C]
nS = len(S)
nC = len(C)
nStudents = nS // nC

score = {1: 100, 2: 50, 3: 20, 4: 0}
ngscore = -1000000


problem = LpProblem("ClassRoomAssignment", LpMaximize)
x = LpVariable.dicts("x", SC, 0, 1, LpBinary)
p = {}
for s in S:
    for c in C:
        p[s, c] = score[int(data_df.loc[s, c])] if data_df.loc[s, c] != 0 else ngscore
# object
problem += lpSum(p[s, c] * x[s, c] for s in S for c in C)


for s in S:
    problem += lpSum(x[s, c] for c in C) == 1

for c in C:
    problem += lpSum(x[s, c] for s in S) >= nStudents
    problem += lpSum(x[s, c] for s in S) <= nStudents + 1

status = problem.solve()
print(f"status: {LpStatus[status]}")
print(f"objective: {pulp.value(problem.objective)}")
print(f"score pre person: {pulp.value(problem.objective) / nS:.4f}")
```

#### Output result


```python
C2S = {}
for c in C:
    C2S[c] = []
    for s in S:
        if x[s, c].value() == 1:
            C2S[c].append(s)


display(C2S)
```

```python
C2Total = {}
C2Score = {(c, i): 0 for c in C for i in range(1, 5)}
for c in C:
    C2Total[c] = 0
    for s in C2S[c]:
        C2Total[c] += p[s, c]
        for i in range(1, 5):
            if data_df.loc[s, c] == i:
                C2Score[c, i] += 1


column = sorted(set([key[0] for key in C2Score.keys()]))
index = [1, 2, 3, 4]

df = pd.DataFrame(index=index, columns=column)

for (c, i), value in C2Score.items():
    df.loc[i, c] = value

df.loc["Satisfaction level"] = [C2Total[c] for c in column]

display(df)
```
